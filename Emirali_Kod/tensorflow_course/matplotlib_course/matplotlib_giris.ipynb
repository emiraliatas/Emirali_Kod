```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpy_array1 = np.linspace(0,10,20)
numpy_array2 = numpy_array1 ** 3
```

```python
(my_figure, my_axises) = plt.subplots(nrows=1, ncols=2)
type(my_axises)

for axis in my_axises:
    axis.plot(numpy_array1,numpy_array2,"g")
    axis.set_xlabel("X Axis")

plt.tight_layout()
```

```python
new_figure = plt.figure()
#dpi = 150
new_axes = new_figure.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_array1, numpy_array2,label="numpy_array1 ** 3")
new_axes.plot(numpy_array1, numpy_array1 ** 2,label="numpy_array1 ** 2")
new_axes.legend(loc=2)
```

```python
new_figure.savefig("newfigure.png", dpi=200)
```
