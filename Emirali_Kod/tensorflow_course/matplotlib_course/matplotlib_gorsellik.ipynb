```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpy_array1 = np.linspace(0,10,20)
numpy_array2 = numpy_array1 ** 2
```

```python
(my_figure, my_axes) = plt.subplots()
my_axes.plot(numpy_array1,numpy_array2,color="#ADBE8B", alpha=0.4)
my_axes.plot(numpy_array2,numpy_array1,color="#1F72A1", alpha=0.7)
```

```python
new_figure, new_axes = plt.subplots()
new_axes.plot(numpy_array1,numpy_array1,color="#1FBE9F")
```
